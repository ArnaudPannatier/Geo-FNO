# NACA theta2mesh

Generate mesh file with function 
```
Theta2Mesh(theta, cnx1 = 50, dy0 = 2.0/120.0, cnx2 = 120, cny = 50, R = 40, Rc = 1.0, L = 40):
```
`theta` is an array with 8 variables

````python
import numpy as np
from scipy import optimize
# symmetrical 4-digit NASA airfoil
# the airfoil is in [0,1]
def NACA_shape(x, digit=12):
    return 5*(digit/100.0)*(0.2969*np.sqrt(x) - 0.1260*x - 0.3516*x**2 + 0.2843*x**3 - 0.1036*x**4)


# generate mesh between a and b 
# dx0, dx0*r, dx0*r^2 ... dx0*r^{N-2}
# b - a = dx0*(r^{N-1} - 1)/(r - 1)
def GeoSpace(a, b, N, r = -1.0, dx0= -1.0):

    xx = np.linspace(a, b, N)

    if r > 1 or dx0 > 0:
        if r > 1: 
            dx0 = (b - a)/((r**(N - 1) - 1)/(r - 1))

            dx = dx0
            for i in range(1, N-1):
                xx[i] = xx[i-1] + dx
                dx *= r
      

        else:
            # first use r=1.05 to generate half of the grids
            # then compute r and generate another half of the grids
            f = lambda r : (r-1)*(b-a) - dx0*(r**(N-1) - 1)
            r = optimize.bisect(f, 1+1e-4, 1.5)
            
            if r > 1.02:
                r = min(r, 1.02)
                dx = dx0
                Nf = 3*N//4
               
                for i in range(1, Nf):
                    xx[i] = xx[i-1] + dx
                    dx *= r

                a = xx[Nf-1]
                dx0 = dx
                
                f = lambda r : (r-1)*(b-a) - dx0*(r**(N - Nf) - 1)
                r = optimize.bisect(f, 1+1e-4, 2.0)
                
                for i in range(Nf, N-1):
                    xx[i] = xx[i-1] + dx
                    dx *= r

            else:
                dx = dx0
                for i in range(1, N-1):
                    xx[i] = xx[i-1] + dx
                    dx *= r


    return xx



# Nx point on top skin
def NACA_shape_mesh(Nx, method="stretching", ratio = 1.0):
    

    if method == "stretching":
        xx = np.zeros(Nx)
        xx[1:] = GeoSpace(0, 1, Nx-1, r =ratio**(1/(Nx-3)))
        xx[1] = xx[2]/4.0 
        

    else:
        print("method : ", method, " is not recognized")


    xx = xx[::-1]
    yy = np.hstack((NACA_shape(xx), -NACA_shape(xx[-2::-1])))
    xx = np.hstack((xx, xx[-2::-1]))

    return xx, yy


# The undeformed box is 
# 0.5 - Lx/2  (8)        0.5 - Lx/6  (7)         0.5 + Lx/6  (6)          0.5 + Lx/2  (5)         (y = Ly/2)
#
# 0.5 - Lx/2  (1)        0.5 - Lx/6  (2)         0.5 + Lx/6  (3)          0.5 + Lx/2  (4)         (y = -Ly/2)
#
# basis function at node (i)   is Bᵢ   = Φᵢ(x) Ψ₁(y)    (1 ≤ i ≤ 4)
# basis function at node (i+4) is Bᵢ₊₄ = Φᵢ(x) Ψ₂(y)    (1 ≤ i ≤ 4)
#
# The map is 
# (x, y) -> (x, y) + dᵢ Bᵢ(x,  y)
#
def NACA_sdesign(d, x, y, Lx=1.5, Ly=0.2):
    x1, x2, x3, x4 = 0.5 - Lx/2,  0.5 - Lx/6,   0.5 + Lx/6,    0.5 + Lx/2
    y1, y2  = - Ly/2,  Ly/2

    
    phi1 = (x - x2)*(x - x3)*(x - x4)  / ((x1 - x2)*(x1 - x3)*(x1 - x4))
    phi2 = (x - x1)*(x - x3)*(x - x4)  / ((x2 - x1)*(x2 - x3)*(x2 - x4))
    phi3 = (x - x1)*(x - x2)*(x - x4)  / ((x3 - x1)*(x3 - x2)*(x3 - x4))
    phi4 = (x - x1)*(x - x2)*(x - x3)  / ((x4 - x1)*(x4 - x2)*(x4 - x3))

    psi1 = (y - y2)/(y1 - y2)
    psi2 = (y - y1)/(y2 - y1)

    B = np.array([phi1*psi1, phi2*psi1,  phi3*psi1,  phi4*psi1,  phi4*psi2, phi3*psi2, phi2*psi2,  phi1*psi2])
    return x, y + np.dot(d, B)







def Cgrid2Cylinder(cnx1, cnx2, cny, Cgrid, Cylinder):
    # Cgrid
    nx1, nx2, ny = cnx1+1, cnx2+1, cny+1
    
    for j in range(cny+1):  
        if j == 0:
            Cylinder[0:cnx1+cnx2, j] = Cgrid[0:cnx1+cnx2]
            Cylinder[cnx1+cnx2:2*nx1+cnx2-1, j] = Cylinder[cnx1::-1, j]
        else:
            Cylinder[:, j] = Cgrid[(j-1)*(2*cnx1+cnx2+1) + cnx1+cnx2: (j-1)*(2*cnx1+cnx2+1) + cnx1+cnx2+1 + 2*cnx1+cnx2]
        

def Cylinder2Cgrid(cnx1, cnx2, cny, Cylinder, Cgrid):
    # Cylinder, 
    nx1, nx2, ny = cnx1+1, cnx2+1, cny+1

    for j in range(cny+1):
        if j == 0:
            Cgrid[0:cnx1+cnx2] = Cylinder[0:cnx1+cnx2, j]
        else:
            Cgrid[(j-1)*(2*cnx1+cnx2+1) + cnx1+cnx2: (j-1)*(2*cnx1+cnx2+1) + cnx1+cnx2+1 + 2*cnx1+cnx2] = Cylinder[:, j]


def PlotCylinderGrid(xx_Cylinder, yy_Cylinder, filename="None", equal_axis = True):
    plt.plot(xx_Cylinder, yy_Cylinder, color = "black", linewidth=0.1)
    plt.plot(xx_Cylinder.T, yy_Cylinder.T, color = "black", linewidth=0.1)

    if equal_axis:
        plt.axis("equal")

    if filename != "None":
        savefig(filename)
        close("all")
    

# c: number of cells
# cnx1 C mesh behind trailing edge 
# cnx2 C mesh around airfoil
# cny radial direction
# 
# The airfoil is in [0,1]
# R: radius of C mesh 
# L: the right end of the mesh
# the bounding box of the mesh is [Rc-R, L], [-R, R]
#
# dy0, vertical mesh size



        
        
def Theta2Mesh(theta, cnx1 = 50, dy0 = 2.0/120.0, cnx2 = 120, cny = 50, R = 40, Rc = 1.0, L = 40):
    assert(len(theta) == 8 and theta[0] == 0.0)

    cnx = 2*cnx1 + cnx2
    nx1, nx2, ny = cnx1+1, cnx2+1, cny+1 #points
    nnodes = (2*nx1+cnx2-1)*cny + (nx1+cnx2-1)

    xx_airfoil, yy_airfoil = NACA_shape_mesh(cnx2//2+1, method="stretching")
    
    
    
    xx_inner = GeoSpace(0, 1, nx1, dx0 = np.sqrt((xx_airfoil[0] - xx_airfoil[1])**2 + (yy_airfoil[0] - yy_airfoil[1])**2)/(L-1))
    xx_outer = GeoSpace(Rc, L, nx1)
    wy = GeoSpace(0, 1, ny, dx0 = dy0/R)
    
    
    
    
    for i in range(cnx2+1):
        xx_airfoil[i], yy_airfoil[i] = NACA_sdesign(theta, xx_airfoil[i], yy_airfoil[i])
    
    
    
    xy_inner, xy_outer = np.zeros((2*nx1+cnx2-1,2)), np.zeros((2*nx1+cnx2-1,2))
    # top flat 
    xy_inner[nx1-1::-1, 0] = xx_airfoil[0]*(1 - xx_inner)  +  L * xx_inner
    xy_inner[nx1-1::-1, 1] = yy_airfoil[0]*(1 - xx_inner)
    xy_outer[0:nx1, 0] = xx_outer[::-1] 
    xy_outer[0:nx1, 1] = R
    
    # airfoil
    xy_inner[nx1-1:nx1+cnx2, 0] = xx_airfoil
    xy_inner[nx1-1:nx1+cnx2, 1] = yy_airfoil

    θθ = np.linspace(np.pi/2, 3*np.pi/2, nx2)
    xy_outer[nx1-1:nx1+cnx2, 0] = R*np.cos(θθ) + Rc
    xy_outer[nx1-1:nx1+cnx2, 1] = R*np.sin(θθ)
    # bottom flat
    xy_inner[nx1+cnx2-1:2*nx1+cnx2-1, 0] = xy_inner[nx1-1::-1, 0]
    xy_inner[nx1+cnx2-1:2*nx1+cnx2-1, 1] = xy_inner[nx1-1::-1, 1]
    xy_outer[nx1+cnx2-1:2*nx1+cnx2-1, 0] = xx_outer 
    xy_outer[nx1+cnx2-1:2*nx1+cnx2-1, 1] = -R

    
    
    # Construct Cylinder grid

    xx_Cylinder = np.zeros((2*nx1+cnx2-1, cny+1))
    yy_Cylinder = np.zeros((2*nx1+cnx2-1, cny+1))
    
    
    for j in range(ny):
        xx_Cylinder[:,j] = xy_inner[:,0]*(1 - wy[j]) + xy_outer[:,0]*wy[j];
        yy_Cylinder[:,j] = xy_inner[:,1]*(1 - wy[j]) + xy_outer[:,1]*wy[j];

    
    return xx_Cylinder, yy_Cylinder
````

Inputs are X coordinate and Y coordinate of the C grid, and pressure on the C grid
Outputs are drag and lift


````python
# visualize pressure coefficent
def visual_Cp(XC, YC, PC, cnx1 = 50, cnx2 = 120, cny = 50):
    gamma = 1.4
    p_oo = 1.0; rho_oo = 1.0; M_oo = 0.8; Aot = 0.0 
    u_oo, v_oo = np.cos(Aot*np.pi/180)*  M_oo*np.sqrt(gamma*p_oo/rho_oo) , np.sin(Aot*np.pi/180) *  M_oo*np.sqrt(gamma*p_oo/rho_oo)

    xx, yy, p = XC[cnx1:-cnx1,0], YC[cnx1:-cnx1,0], PC[cnx1:-cnx1,0]
    C_p = (p - p_oo)/(0.5*rho_oo*u_oo**2)
    
    
    plt.figure(figsize=(16, 8))
    plt.plot(xx[0:cnx2//2+1], -C_p[0:cnx2//2+1], "-", label = "top")
    plt.plot(xx[cnx2//2:], -C_p[cnx2//2:], "-", label = "bottom")
    plt.ylabel("- Cp")
    plt.xlabel("x")
    plt.legend()
    
    
    plt.figure(figsize=(16, 8))
    n = 40
    plt.pcolormesh(XC[cnx1-n:-cnx1+n,0:n], YC[cnx1-n:-cnx1+n,0:n], PC[cnx1-n:-cnx1+n,0:n], shading='gouraud')
    plt.colorbar()
    plt.axis('equal')

# compute drag and lift
def compute_F(XC, YC, p, cnx1 = 50, cnx2 = 120, cny = 50):
    xx, yy, p = XC[cnx1:-cnx1,0], YC[cnx1:-cnx1,0], p[cnx1:-cnx1,0]
    
    F = np.array([0.0, 0.0])
    for i in range(cnx2):
        vn = np.array([(yy[i+1] - yy[i]), -(xx[i+1] - xx[i])])
        F -= vn * (p[i] + p[i + 1])/2.0
    
    return F
    
ind = 20
visual_Cp(XC[ind, :, :], YC[ind, :, :], QC[ind, 3, :, :])
F = compute_Cd(XC[ind, :, :], YC[ind, :, :], QC[ind, 3, :, :])
print("F = ", F)
````
